```python
import numpy as np

def check_valid_pdf(num_outcomes, values, probabilities):
    if num_outcomes != len(values) or num_outcomes != len(probabilities):
        return False, "Number of outcomes, values, and probabilities must be consistent."
    if sum(probabilities) == 1:
        return True, "It is Valid PDF."
    else:
        return False, "Probabilities do not sum up to 1 PDF Should be 1."

def calculate_mean(values, probabilities):
    return np.dot(values, probabilities)

def calculate_variance(values, probabilities, mean):
    return np.dot(probabilities, (values - mean)**2)
  

def m():
  try:
    num_outcomes = int(input("Enter the number of possible outcomes: "))
    assert(num_outcomes > 0), 'Number must be bigger than 0'
  except:
    m()
  try:
    values = [float(val) for val in input("Enter values by space: ").split(' ')]
  except:
    m()

  try:
    probabilities = [float(prob) for prob in input("Enter probabilities by space: ").split(' ')]
  except:
    m()


  is_valid, message = check_valid_pdf(num_outcomes, values, probabilities)
  mean = calculate_mean(values, probabilities)
  variance = calculate_variance(values, probabilities, mean)
  if is_valid == True:
    print(message)
    print(f"Mean is: {mean}")
    print(f"Variance is: {variance}")
  else:
    print(message)
    print('-------------------------------------------------------------')
    m()

m()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_coin_toss(num_tosses, probability):
    outcomes = ['Heads' if np.random.binomial(1, probability) == 1 else 'Tails' for _ in range(num_tosses)]
    return outcomes

def calculate_relative_frequencies(outcomes):
    num_heads = outcomes.count('Heads')
    num_tails = outcomes.count('Tails')
    relative_frequency_heads = num_heads / len(outcomes)
    relative_frequency_tails = num_tails / len(outcomes)
    return relative_frequency_heads, relative_frequency_tails

def plot_relative_frequencies(relative_frequencies_heads, relative_frequencies_tails):
    plt.plot(relative_frequencies_heads, label='Heads')
    plt.axhline(0.5, color='red', linestyle='--', label='Theoretical Probability (0.5)')
    plt.xlabel('Number of Tosses')
    plt.ylabel('Relative Frequency')
    plt.legend()
    plt.title('Law of Large Numbers - Coin Toss Simulation')
    plt.show()

num_tosses = int(input("Enter the number of coins to toss:"))
probability = 0.5
outcomes = simulate_coin_toss(num_tosses, probability)
relative_frequencies_heads, relative_frequencies_tails = zip(*[calculate_relative_frequencies(outcomes[:i+1]) for i in range(num_tosses)])
plot_relative_frequencies(relative_frequencies_heads, relative_frequencies_tails)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_dice_roll(num_rolls):
    """Simulates rolling a 6-sided die `num_rolls` times and returns the outcomes as a NumPy array."""
    return np.random.choice(range(1, 7), size=num_rolls)

def calculate_relative_frequencies(outcomes):
    """Calculates the relative frequencies of each side occurring in the outcomes."""
    sides = np.unique(outcomes)  # Get unique sides
    counts = np.bincount(outcomes - 1)  # Count occurrences efficiently
    return counts / len(outcomes)  # Calculate relative frequencies

def plot_relative_frequencies(num_rolls):
    """Simulates rolling a die `num_rolls` times, calculates relative frequencies, and plots them."""
    sides = 6
    outcomes = simulate_dice_roll(num_rolls)
    relative_frequencies = calculate_relative_frequencies(outcomes)

    plt.plot(relative_frequencies, label='Relative Frequency')
    plt.axhline(1 / sides, color='red', linestyle='--', label='Theoretical Probability')
    plt.xlabel('Number of Rolls')
    plt.ylabel('Relative Frequency')
    plt.legend()
    plt.title(f'Law of Large Numbers - Dice Roll Simulation ({num_rolls} rolls)')
    plt.show()

# Get number of rolls from user
num_rolls = int(input("Enter the number of dice to roll: "))

# Simulate, calculate, and plot
plot_relative_frequencies(num_rolls)
```

```python
def findmean(probability_list, x_random_variable, object_list):
  mean = 0
  for i in range(len(probability_list)):
      mean += object_list[i] * probability_list[i]

  return mean


def findvar(probability_list, x_random_variable, object_list):
  mean = findmean(probability_list, x_random_variable, object_list)
  variance = 0
  for i in range(len(probability_list)):
      variance += (object_list[i] - mean)**2 * probability_list[i]

  return variance


import numpy as np

# x_random_variable = 5
# object_list = [0, 1, 2, 3, 4]
# probability_list = [0.6561, 0.2916, 0.0486, 0.0036, 0.0001]
x_random_variable = int(input("Enter the Number of Possible Variables: "))
object_list = [float(val) for val in input("Enter Values: ").split(',')]
probability_list = [float(prob) for prob in input("Enter Probability: ").split(',')]

if any(p < 0 for p in probability_list):
    print("Invalid Probability Distribution: probability values cannot be negative")
elif np.sum(probability_list) != 1:
    print("Invalid Probability Distribution: probabilities must sum to 1")
else:
    mean = findmean(probability_list, x_random_variable, object_list)
    variance = findvar(probability_list, x_random_variable, object_list)

    print("Mean:", mean)
    print("Variance:", variance)

```

```python
import math
def nCr(n,r):
    if r>n:
        return 0
    if r==0 and n==r:
        return 1
    res=0
    for i in range(r):
        res+=math.log(n-i) - math.log(i+1)
    return round(math.exp(res))
```

```python
def rangeofx(n,k,N):
    x = []
    for i in range(max(0,n+k-N), max(k,n)):
        x.append(i)
    return x
```

```python
def eqnHypergeometric(k,x,N,n):
    return (nCr(k,x)*nCr(N-k, n-x) / nCr(N,n))
```

```python
N = int(input("Enter the value of N: "))
n = int(input("Enter the value of n: "))
k = int(input("Enter the value of k: "))

x = rangeofx(n,k,N)
pdf=[]
for i in x:
    p_of_x = eqnHypergeometric(k,x[i],N,n)
    pdf.append(p_of_x)

print(pdf)
```

```python
import matplotlib.pyplot as plt

outcomes = range(len(pdf))  

plt.bar(outcomes, pdf)

plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("PDF of Hypergeometric Distribution")


plt.show()
```

```python
from scipy import integrate

def integrate_function(func, lower_limit, upper_limit):
    """
    Computes the definite integral of a given function.

    Parameters:
        func (callable): The function to be integrated.
        lower_limit (float): Lower limit of integration.
        upper_limit (float): Upper limit of integration.

    Returns:
        float: The definite integral of the function over the given limits.
    """
    result, _ = integrate.quad(func, lower_limit, upper_limit)
    return result

# Example usage:
def example_function(x):
    return x ** 2  # Example function: x^2

lower_limit = 0
upper_limit = 1
result = integrate_function(example_function, lower_limit, upper_limit)
print("Definite integral:", result)
```

```python
import math
import matplotlib.pyplot as plt

def poisson_distribution(lambda_, n):
    # Calculate the Poisson probability mass function
    probability = (math.exp(-lambda_) * lambda_**n) / math.factorial(n)
    return probability

def poisson_cdf(lambda_, max_n):
    cdf_values = []
    cumulative_probability = 0
    for n in range(max_n + 1):
        probability = poisson_distribution(lambda_, n)
        cumulative_probability += probability
        cdf_values.append(cumulative_probability)
    return cdf_values

def plot_poisson_distribution(lambda_, max_n):
    n_values = list(range(max_n + 1))
    probabilities = [poisson_distribution(lambda_, n) for n in n_values]
    cdf_values = poisson_cdf(lambda_, max_n)

    plt.figure(figsize=(12, 5))

    # Plot PMF
    plt.subplot(1, 2, 1)
    plt.bar(n_values, probabilities, color='lightblue')
    plt.xlabel('n')
    plt.ylabel('Probability')
    plt.title(f'PMF of Poisson Distribution for λ={lambda_}')
    plt.grid(True)

    # Plot CDF
    plt.subplot(1, 2, 2)
    plt.plot(n_values, cdf_values, marker='o', linestyle='-')
    plt.xlabel('n')
    plt.ylabel('Cumulative Probability')
    plt.title(f'CDF of Poisson Distribution for λ={lambda_}')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def main():
    # Get input from the user for lambda and n
    lambda_ = float(input("Enter the value of lambda (λ): "))
    max_n = int(input("Enter the maximum value of n for the graph: "))

    # Calculate mean and variance
    mean = lambda_
    variance = lambda_

    print(f"Mean: {mean}")
    print(f"Variance: {variance}")

    # Plot the Poisson distribution and CDF
    plot_poisson_distribution(lambda_, max_n)

if __name__ == "__main__":
    main()
```
